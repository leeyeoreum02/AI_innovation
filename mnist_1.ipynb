```
%tensorflow_version 1.x

import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np
from datetime import datetime


mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)

print()
print("train.num = ", mnist.train.num_examples,
      ", test.num = ", mnist.test.num_examples,
      ", validation.num = ", mnist.validation.num_examples)
```

```
print("type(mnist) = ", type(mnist), ", type(mnist.train.images) = ", type(mnist.train.images), ", type(mnist.train.labels) = ", type(mnist.train.labels))
print()
print("mnist.train.images.shape = ", mnist.train.images.shape, "np.shape(mnist.train.images) = ", np.shape(mnist.train.images))
```

```
learning_rate = 1e-1
epochs = 100
batch_size = 100

input_nodes = 784
hidden_nodes = 100
output_nodes = 10
```

```
X = tf.placeholder(tf.float32, [None, 784])  # 8개 입력노드
T = tf.placeholder(tf.float32, [None, 10])  # 1개 정답노드

W2 = tf.Variable(tf.random_normal([input_nodes, hidden_nodes]))  # 가중치 노드
b2 = tf.Variable(tf.random_normal([hidden_nodes]))     # 바이어스 노드

W3 = tf.Variable(tf.random_normal([hidden_nodes, output_nodes]))  # 가중치 노드
b3 = tf.Variable(tf.random_normal([output_nodes]))     # 바이어스 노드
```

```
Z2 = tf.matmul(X, W2) + b2
A2 = tf.nn.relu(Z2)

Z3 = logits = tf.matmul(A2, W3) + b3

y = A3 = tf.nn.softmax(Z3)
```

```
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=Z3, labels=T))

optimizer = tf.train.GradientDescentOptimizer(learning_rate)

train = optimizer.minimize(loss)
```

```
predicted_val = tf.equal(tf.argmax(A3, 1), tf.argmax(T, 1))

accuracy = tf.reduce_mean(tf.cast(predicted_val, dtype=tf.float32))

accuracy_index = tf.cast(predicted_val, dtype=tf.float32)

predicted_list = tf.argmax(A3, 1)
```

```
with tf.Session() as sess:

  sess.run(tf.global_variables_initializer())

  start_time = datetime.now()

  for i in range(epochs):
    
    total_batch = int(mnist.train.num_examples / batch_size)

    for step in range(total_batch):

      batch_x_data, batch_t_data = mnist.train.next_batch(batch_size)

      loss_val, _ = sess.run([loss, train], feed_dict={X: batch_x_data, T: batch_t_data})

      if step % 100 == 0:
        print("===============================================================")
        print("step = ", step, ", loss_val = ", loss_val)
        print("accuracy_index = ", accuracy_index)
        print("predicted_list = ", predicted_list)
        print("===============================================================")

  test_x_data = mnist.test.images
  test_t_data = mnist.test.labels

  accuracy_val = sess.run(accuracy, feed_dict={X: test_x_data, T: test_t_data})

  print()
  print("Accuracy = ", accuracy_val)

end_time = datetime.now() 
print()
print("Elapsed time = ", end_time - start_time) 
```
