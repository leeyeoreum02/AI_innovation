```
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD

import numpy as np

print(tf.__version__)
```

```
x_data = np.array([1, 2, 0, 5, 4, 3, 1, 2, -1, 3, 1, 0, 2, 4, 2, 4, 1, 2, -1, 3, 2, 4, 3, 3, 0, 2, 6, 2, 2, 1, 1, -2, -2, 0, 1, 3, 1, 1, 3, 0, 1, 4, 2, 3, 3]).reshape(-1, 3)
t_data = np.array([-4, 4, -6, 3, -4, 9, -7, 5, 6, 0, 4, 3, 5, 5, 1]).reshape(-1, 1)

print("x_data.shape = ", x_data.shape, ", x_data.ndim = ", x_data.ndim)
print("t_data.shape = ", t_data.shape, ", t_data.ndim = ", t_data.ndim)
```

```
model = Sequential()

model.add(Flatten(input_shape=(x_data.shape[1],)))

model.add(Dense(t_data.shape[1], activation='linear'))
```

```
model.compile(optimizer=SGD(learning_rate=1e-2), loss='mse')

model.summary()
```

```
hist = model.fit(x_data, t_data, epochs=1000)
```

```
type(hist)
hist.history.keys()
```

```
result = model.predict(np.array([5, 5, 0, 2, 3, 1, -1, 0, -1, 10, 5, 2, 4, -1, -2]).reshape(-1, 3))

print("type(result) = ", type(result), '\n')
print(result)
```
