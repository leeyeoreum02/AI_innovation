```
import tensorflow as tf

tf.__version__
```

```
a = tf.constant(1.0)
b = tf.constant(2.0)

c = a + b

print(c.numpy())
```

```
W = tf.Variable(tf.random.normal([1]))
print('initial W = ', W.numpy())
print('====================================================')

for step in range(2):
  W = W + 1.0
  print('step = ', step, ', W = ', W.numpy())
```

```
a = tf.constant(1.0)
b = tf.constant(3.0)

def tensor_sum(x, y):
  return x + y

result = tensor_sum(a, b)

print(type(result))
print(result.numpy())
```
